# Missing Value Analysis and Treatment :

```python
datasets = {
    "customers": customers,
    "orders": orders,
    "products": products,
    "payments": payments,
    "reviews": reviews,
    "items": items,
    "sellers": sellers,
    "geo": geo,
    "translation": translation
}

for name, df in datasets.items():
    print("="*40)
    print("Dataset:", name)
    print(df.isnull().sum())
    print()
```

### Observation :

Missing value analysis was performed across all nine datasets to identify incomplete records and determine the appropriate treatment strategy.

Instead of filling every missing value blindly, each column was evaluated based on its business meaning.

### For Orders :

```python
orders[orders["order_status"] != "delivered"]["order_status"].value_counts()
```

### Cleaning Decision :

Delivery-related columns contain missing values because some orders were canceled, unavailable, or still in processing.

These missing values represent valid business scenarios rather than data quality issues. Therefore, they were intentionally retained.

### For Products :

```python
# product_category_name :  

products["product_category_name"] = products[
    "product_category_name"
].fillna("Unknown")

#Product should not have blank category
```

```python
# Length/photo columns : (these are meta-data, fill them w median)

products["product_name_lenght"] = products[
    "product_name_lenght"
].fillna(
    products["product_name_lenght"].median()
)

products["product_description_lenght"] = products[
    "product_description_lenght"
].fillna(
    products["product_description_lenght"].median()
)

products["product_photos_qty"] = products[
    "product_photos_qty"
].fillna(
    products["product_photos_qty"].median()
)
```

```python
# Weight/size columns : fill median

products["product_weight_g"] = products[
    "product_weight_g"
].fillna(
    products["product_weight_g"].median()
)

products["product_length_cm"] = products[
    "product_length_cm"
].fillna(
    products["product_length_cm"].median()
)

products["product_height_cm"] = products[
    "product_height_cm"
].fillna(
    products["product_height_cm"].median()
)

products["product_width_cm"] = products[
    "product_width_cm"
].fillna(
    products["product_width_cm"].median()
)
```

```python
products.isnull().sum()
```

### Observation :

Missing values in the products dataset were primarily found in product category and product metadata columns.

Categorical missing values were replaced with **"Unknown"**, while numerical attributes such as dimensions and product measurements were imputed using the median to preserve the overall distribution.

### For Reviews :

```python
reviews["review_comment_message"] = reviews[
    "review_comment_message"
].fillna("No Comment")
```

```python
reviews["review_comment_title"] = reviews[
    "review_comment_title"
].fillna("No Comment")
```

### Cleaning Decision :

A large number of customers provided ratings without leaving written comments.

Since review comments are optional, these missing values were preserved because they represent genuine customer behavior rather than incomplete data.

## Feature Engineering :

```python
orders["purchase_month"] = orders[
    "order_purchase_timestamp"
].dt.month
```

```python
orders["purchase_month"].head()
```

```python
orders["purchase_year"] = orders[
    "order_purchase_timestamp"
].dt.year
```

```python
orders["order_estimated_delivery_date"] = pd.to_datetime(
    orders["order_estimated_delivery_date"]
)
```

```python
orders.info()
```

```python
orders["delivery_delay"] = (
    orders["order_delivered_customer_date"] -
    orders["order_estimated_delivery_date"]
).dt.days
```

```python
orders["delivery_delay"].describe()
```

### Observation :

New features such as purchase month, purchase year, delivery time, and delivery delay were created to support trend analysis and business insights.

These engineered features will also improve dashboard creation and SQL analysis.
